```python
%tensorflow_version 1.x
!pip install gpt-2-simple
import gpt_2_simple as gpt2
```

```python
gpt2.download_gpt2(model_name="124M")
```

```python
file_name="Kanye_West.txt"
```

```python
sess=gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='124M',
              steps=200,
              restore_from='fresh',
              run_name='run1',
              print_every=10,
              sample_every=20,
              save_every=20
              )
```

```python
gpt2.generate(sess,run_name='run1')
```

```python
gpt2.generate(sess,
              length=10,
              temperature=0.2,
              prefix="I love deep learning",
              nsamples=5,
              batch_size=5,
              top_k=40
              )
```
